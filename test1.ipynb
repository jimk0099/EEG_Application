## Import the few packages we'll need

```python
import matplotlib
import pathlib
import mne

import pandas as pd
import numpy as np

matplotlib.use('Qt5Agg')
```

```python
# source: https://mne.discourse.group/t/reading-from-csv-file/3458

User_frame = pd.read_csv("/home/jimk/Documents/NTUA/semester10/kiniti_ygeia/eeg_application/data/BrainFlow-RAW_test2-right-punch_0.csv", sep='\t', decimal='.')
User_frame = User_frame.apply(lambda x: x.str.replace(',','.'))

# assigning the channel type when initializing the Info object
ch_names = ['Sample Index', 'EXG Channel 0', 'EXG Channel 1', 'EXG Channel 2', 'EXG Channel 3', 
'EXG Channel 4', 'EXG Channel 5', 'EXG Channel 6', 'EXG Channel 7', 'EXG Channel 8', 'EXG Channel 9', 
'EXG Channel 10', 'EXG Channel 11', 'EXG Channel 12', 'EXG Channel 13', 'EXG Channel 14', 'EXG Channel 15', 
'Accel Channel 0', 'Accel Channel 1', 'Accel Channel 2', 'Other', 'Other', 'Other', 'Other', 'Other', 
'Other', 'Other', 'Analog Channel 0', 'Analog Channel 1', 'Analog Channel 2', 'Timestamp', 'Other']

ch_types = ['misc', 'eeg', 'eeg', 'eeg', 'eeg', 
            'eeg', 'eeg', 'eeg', 'eeg', 'eeg', 'eeg', 
            'eeg', 'eeg', 'eeg', 'eeg', 'eeg', 'eeg',
            'misc', 'misc', 'misc', 'misc', 'misc', 'misc', 'misc', 'misc',  
            'misc', 'misc', 'misc', 'misc', 'misc', 'misc', 'misc']

User_frame.columns = ch_names
User_frame.head()
```

```python
#data= pd.DataFrame.to_numpy(User_frame, dtype= np.float64)
data = User_frame.transpose().to_numpy()
```

Note: We do not have the last column ('Timestamp (Formatted)') in the csv for some reason

```python
sampling_freq = 125  # in Hertz

#info = mne.create_info(ch_names= ch_names, sfreq= sampling_freq)
info = mne.create_info(ch_names= ch_names, ch_types= ch_types, sfreq= sampling_freq)

User_raw = mne.io.RawArray(data, info)

User_raw
```
